Задачи:
* Подобрать оптимальные наборы признаков для обучения модели линейной регрессии и случайного леса. Избавится от мультиколлинеарности.
* Обучить модели с различными параметрами и наборами признаков.

### Предобработка

Введу след названия для наборов признаков
* Набор без изменений - s1
* Наборы очищенные от кореллирующих признаков без дополнительных признаков буду называть с c(c1..cn).
* Наборы с заменой средних частот посещения зала в мес. и общ. на относительную последнего месяца буду называть с a(а1..an).
* Наборы с заменой периода до конца абонемента на относительный период до конца абонемента буду называть с b(b1..bn)
* Наборы с признаком созданным из лайфтайма и длит. абонемента с d(d1..dn)


#### Наборы признаков

Для обучения моделей на новом наборе необходимо описать его ниже из начального df_s1 добавить в лист feature_sets_list и вписать в мультииндекс results_df.

```python
y = df['churn']

#Параметр random_state у моделей 2352112 или None
r_seed = 3
```

```python
results_df = pd.DataFrame(columns=['accuracy', 'precision', 'recall'], 
                          index= pd.MultiIndex.from_tuples(itertools.product(('LR', 'RF'),('s1', 'c1', 'c2','c3', 'a1', 'b1', 'a1b1', 'c3a1b1'))))
```

```python
feature_sets_list = list()
```

Набор s1

```python
df_s1 = df.drop('churn', axis=1)
feature_sets_list.append(df_s1)
```

Набор c1, c2

```python
df_c1 = df_s1.drop(['month_to_end_contract', 'avg_class_frequency_total'],  axis=1)
df_c2 = df_s1.drop(['contract_period', 'avg_class_frequency_total'], axis=1)
feature_sets_list.append(df_c1)
feature_sets_list.append(df_c2)
```

Набор с3

```python
df_c3 = df_s1.drop(['age', 'phone',],  axis=1)
feature_sets_list.append(df_c3)
```

Набор a1

```python
df_a1 = df_s1.copy()
df_a1['relative_avg_class_frequency_current_month'] = (
    df_a1['avg_class_frequency_current_month'] - 
    df_a1['avg_class_frequency_total']
)
df_a1.drop(['avg_class_frequency_current_month', 'avg_class_frequency_total'], axis = 1, inplace=True)
feature_sets_list.append(df_a1)
```

Набор b1

```python
df_b1 = df_s1.copy()
df_b1['relative_period_to_end_contract'] = (
    df_b1['month_to_end_contract'] /
    df_b1['contract_period']
)
df_b1.drop('month_to_end_contract', axis=1, inplace=True)

feature_sets_list.append(df_b1)
```

Набор a1b1

```python
df_a1b1 = df_a1.copy()
df_a1b1['relative_period_to_end_contract'] = (
    df_a1b1['month_to_end_contract'] /
    df_a1b1['contract_period']
)
df_a1b1.drop('month_to_end_contract', axis=1, inplace=True)
feature_sets_list.append(df_a1b1)
```

Набор c3a1b1

```python
df_c3a1b1 = df_a1b1.copy()
df_c3a1b1.drop(['age', 'phone',],  axis=1, inplace=True)
feature_sets_list.append(df_c3a1b1)
```

#### Разбиение на обучающую и валидационную выборки

```python
X_train, X_test, y_train, y_test =\
np.transpose([train_test_split(f_s, y, test_size = 0.2, random_state = r_seed) for f_s in feature_sets_list])
```

#### Стандартизация

```python
scaler = StandardScaler()

for i in range(len(X_test)):
    X_test[i] = scaler.fit_transform(X_test[i])
    X_train[i] = scaler.transform(X_train[i])
```

### Обучение

#### Логистическая регрессия

```python
lr_predictions_list = list()
lr_predict_proba_list = list()
```

```python
#Обучаю по умолчанию с l2 регуляризацией
for i in range(len(X_train)):
    model = LogisticRegression(random_state = r_seed).fit(X_train[i], y_train[i])
    lr_predictions_list.append(model.predict(X_test[i]))
    lr_predict_proba_list.append(model.predict_proba(X_test[i]))
```

#### Случайный лес

```python
rf_predictions_list = list()
rf_predict_proba_list = list()
```

```python
for i in range(len(X_train)):
    model = RandomForestClassifier(random_state = r_seed).fit(X_train[i], y_train[i])
    rf_predictions_list.append(model.predict(X_test[i]))
    rf_predict_proba_list.append(model.predict_proba(X_test[i]))
```

### Оценка

```python
#Если истинно, предсказания модели заменяюттся на предсказания с кастомным порогом.
custom_balance = True
threshold = 0.3
if custom_balance :
    for i in range(len(lr_predictions_list)):
        lr_predictions_list[i] = [0 if p < threshold else 1 for p in lr_predict_proba_list[i][:,1]]
    for i in range(len(rf_predictions_list)):
        rf_predictions_list[i] = [0 if p < threshold else 1 for p in rf_predict_proba_list[i][:,1]]
```

```python
metrics = ['accuracy', 'precision', 'recall']

results_df.loc['LR','accuracy'] = [accuracy_score(y_test[i], lr_predictions_list[i]) for i in range(len(y_test))]
results_df.loc['RF','accuracy'] = [accuracy_score(y_test[i], rf_predictions_list[i]) for i in range(len(y_test))]
results_df.loc['LR','precision'] = [precision_score(y_test[i], lr_predictions_list[i]) for i in range(len(y_test))]
results_df.loc['RF','precision'] = [precision_score(y_test[i], rf_predictions_list[i]) for i in range(len(y_test))]
results_df.loc['LR','recall'] = [recall_score(y_test[i], lr_predictions_list[i]) for i in range(len(y_test))]
results_df.loc['RF','recall'] = [recall_score(y_test[i], rf_predictions_list[i]) for i in range(len(y_test))]
```

Основные принципы стратегии удержания :

* В зависимости от основного кластера клиента(до прогноза оттока) предпринимаем различные действия:


     * 0 'Мутные ребята' : Хорошо бы получить отзыв от этой категории клиентов и понять почему не один из них не оставил номера телефона. Если все абонементы окажутся подаренными я не удивлюсь.   
       
     * 1 'Пришел-оттек' : Основной кластер оттока, у большинства в этом кластере множество негативных признаков склоняющих к попаданию в отток. Так же большинство новички в зале. В нем явно есть те, кто уйдут в любом случае. Однако, так-же в этот кластер попадают пользователи с большим lifetime, которые давно склонялись к попаданию в отток и не получили должного внимания. В этом кластере важно смотреть на lifetime клиента.
     
     * 2 'Узники дружбы' Думаю для этой категории клиентов все зависит от партнера. Уйдет партнер, уйдет и клиент. Групповые занятия могут решать проблему в каких то случаях.

     * 4 'Золотая середина' Требуют наибольшего внимания, из них формируется костяк зала  или они пополняют ряды 1 кластера. Основной таргет для промо-акций и прочих поощрений.
