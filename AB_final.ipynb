```python
cl_product_page_pivot =\
pd.pivot_table(t_participants_df[t_participants_df['event_name'] == 'product_page'],
               index=['first_date', 'group'],
               columns='lifetime',
               values='user_id',
               aggfunc='count').fillna(0).cumsum(axis=1)

cl_product_cart_pivot =\
pd.pivot_table(t_participants_df[t_participants_df['event_name'] == 'product_cart'],
               index=['first_date', 'group'],
               columns='lifetime',
               values='user_id',
               aggfunc='count').fillna(0).cumsum(axis=1)

cl_purchase_pivot =\
pd.pivot_table(t_participants_df[t_participants_df['event_name'] == 'purchase'],
               index=['first_date', 'group'],
               columns='lifetime',
               values='user_id',
               aggfunc='count').fillna(0).cumsum(axis=1)
```

###  –î–∏–Ω–∞–º–∏–∫–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –≤ –≤–æ—Ä–æ–Ω–∫–µ –Ω–∞ —Ä–∞–∑–Ω—ã—Ö —ç—Ç–∞–ø–∞—Ö –≤ —Å—Ä–µ–∑–µ –≤—ã–±–æ—Ä–æ–∫

#### –î–∏–Ω–∞–º–∏–∫–∞ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –≤ –ø—Ä–æ—Å–º–æ—Ç—Ä –∫–∞—Ä—Ç–æ—á–µ–∫ —Ç–æ–≤–∞—Ä–æ–≤ - product_page

#### –ö—É–º—É–ª—è—Ç–∏–≤–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫:

```python
tmp = t_participants_df[t_participants_df['event_name'] == 'product_page']

tmp = tmp.groupby(['group',tmp['event_dt'].dt.date])['user_id'].count().reset_index()
tmp = tmp.rename({'user_id':'product_page_count'}, axis='columns')

#–†–∞—Å—á–∏—Ç—ã–≤–∞—é –∫—É–º—É–ª—è—Ç–∏–≤–Ω—É—é –∫–æ–Ω–≤–µ—Ä—Å–∏—é –≤ –ø—Ä–æ—Å–º–æ—Ç—Ä—ã –∫–∞—Ä—Ç–æ—á–µ–∫ —Ç–æ–≤–∞—Ä–∞
tmp['q_product_page_count'] = tmp.groupby('group')['product_page_count'].cumsum()

#–†–∞—Å—á–∏—Ç—ã–≤–∞—é –∫–æ–º–º—É–ª—è—Ç–∏–≤–Ω—É—é –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—É—é –∫–æ–Ω–≤–µ—Ä—Å–∏—é —Å–æ–≥–ª–∞—Å–Ω–æ –∏–º–µ—é—â–∏–º—Å—è –≤ —Å–∏—Å—Ç–µ–º–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º –Ω–∞ –∫–∞–∂–¥—É—é –¥–∞—Ç—É.
#–ö–æ–≥–æ—Ä—Ç—ã –ø–æ –≥—Ä—É–ø–ø–∞–º –∏ –¥–∞—Ç–∞–º
u_ = t_participants_df.groupby(['group', t_participants_df['first_date'].dt.date])['user_id'].nunique().reset_index().rename({'user_id':'user_count', 'first_date':'event_dt'}, axis='columns')
#–î–æ–±–∞–≤–ª—è—é –≤ –æ—Å–Ω–æ–≤–Ω–æ–π —Ñ—Ä–µ–π–º —á—Ç–æ–±—ã –ø–æ—Å—á–∏—Ç–∞—Ç—å —á–∏—Å–ª–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–ª—è –∫–∞–∂–¥–æ–π –¥–∞—Ç—ã –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ
tmp = tmp.merge(u_, on=['group', 'event_dt'], how='left').fillna(0)
tmp['q_user_count'] = tmp.groupby('group')['user_count'].cumsum()
tmp['q_conv'] = tmp['q_product_page_count']/tmp['q_user_count']
```

```python
#tmp
```

```python
_=plt.figure(figsize = (16, 9))


plt.title('–ö—É–º—É–ª—è—Ç–∏–≤–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏')
plt.xlabel('–î–∞—Ç–∞')
plt.ylabel('–ö–æ–Ω–≤–µ—Ä—Å–∏—è')
_=sns.lineplot(x='event_dt', y='q_conv', data=tmp, hue='group',  hue_order = ['A', 'B'])
```

```python
alpha = .05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

successes = np.array([sa, sb])
trials = np.array([ta, tb])

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ:
p1 = successes[0]/trials[0]

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤–æ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø–µ:
p2 = successes[1]/trials[1]

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# —Ä–∞–∑–Ω–∏—Ü–∞ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö
difference = p1 - p2 

# —Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –≤ —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è—Ö —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# –∑–∞–¥–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (—Å—Ä–µ–¥–Ω–µ–µ 0, —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 1)
distr = st.norm(0, 1)  
```

```python
alpha = .05 # –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

successes = np.array([sa, sb])
trials = np.array([ta, tb])

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –ø–µ—Ä–≤–æ–π –≥—Ä—É–ø–ø–µ:
p1 = successes[0]/trials[0]

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤–æ –≤—Ç–æ—Ä–æ–π –≥—Ä—É–ø–ø–µ:
p2 = successes[1]/trials[1]

# –ø—Ä–æ–ø–æ—Ä—Ü–∏—è —É—Å–ø–µ—Ö–æ–≤ –≤ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# —Ä–∞–∑–Ω–∏—Ü–∞ –ø—Ä–æ–ø–æ—Ä—Ü–∏–π –≤ –¥–∞—Ç–∞—Å–µ—Ç–∞—Ö
difference = p1 - p2 

# —Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –≤ —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è—Ö —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# –∑–∞–¥–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (—Å—Ä–µ–¥–Ω–µ–µ 0, —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 1)
distr = st.norm(0, 1)  
```

```python
# —Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –≤ —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è—Ö —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
z_value = difference / mth.sqrt(
    p_combined * (1 - p_combined) * (1 / trials[0] + 1 / trials[1])
)

# –∑–∞–¥–∞–µ–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (—Å—Ä–µ–¥–Ω–µ–µ 0, —Å—Ç.–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ 1)
distr = st.norm(0, 1)

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ: ', p_value)

if p_value < alpha:
    print('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: –º–µ–∂–¥—É –¥–æ–ª—è–º–∏ –µ—Å—Ç—å –∑–Ω–∞—á–∏–º–∞—è —Ä–∞–∑–Ω–∏—Ü–∞')
else:
    print(
        '–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –Ω–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π —Å—á–∏—Ç–∞—Ç—å –¥–æ–ª–∏ —Ä–∞–∑–Ω—ã–º–∏'
    )
```
