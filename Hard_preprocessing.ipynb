P.S. –≤—Ä–æ–¥–µ –Ω–∏–∂–µ —Ç—ã —á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –∏ –¥–µ–ª–∞–µ—à—å :)
</div>    

```python
#—Ç–æ–ª—å–∫–æ —Å–µ—Ç–µ–≤—ã–µ –∑–∞–≤–µ–¥–µ–Ω–∏—è
df_chain = df[df['chain'] == '–¥–∞']
#–ó–¥–µ—Å—å –≤—Å–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Å–µ—Ç–µ–≤—ã—Ö –∑–∞–≤–µ–¥–µ–Ω–∏–π, –∏—Ö —è –±—É–¥—É –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞—Ç—å –∏ —Å—á–∏—Ç–∞—Ç—å —á–∞—Å—Ç–æ—Ç—É –≤—Å—Ç—Ä–µ—á–∏.
all_chain_names = df_chain['object_name']
#–ó–¥–µ—Å—å —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –Ω–∞–∑–≤–∞–Ω–∏—è –≤ –ø–µ—Ä–≤–æ–º —Å—Ç–æ–ª–±—Ü–µ –∏ –≤–æ –≤—Ç–æ—Ä–æ–º –æ–Ω–∏ –∂–µ –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏
tmp = pd.Series(df_chain['object_name'].unique())
tmp = pd.concat([tmp, tmp.str.lower()], axis=1)
```

#### –ü–µ—Ä–≤—ã–π —ç—Ç–∞–ø

–û—Ç—Ñ–∏–ª—å—Ç—Ä–æ–≤—ã–≤–∞—é –≤—Å–µ –æ–±—â–∏–µ —Å–ª–æ–≤–∞

```python
#–°–ª–æ–≤–∞ –Ω–∞ –≤—ã–±—Ä–æ—Å
bad_words = df['object_type'].unique().tolist()
bad_words.extend(['–ø–∏—Ü—Ü–µ—Ä–∏—è','–ø–∏—Ü—Ü–∞', '–∫–∞–ª—å—è–Ω–Ω–∞—è', '–∫–∞–ª—å—è–Ω', '–ª–∞—É–Ω–¥–∂',\
                  '–ø–µ–∫–∞—Ä–Ω—è', '–∫–æ–Ω–¥–∏—Ç–µ—Ä—Å–∫–∞—è', '–∫–æ—Ñ–µ–π–Ω—è', '–∫–∞—Ñ—Ñ–µ','–∞–Ω—Ç–∏', '–∏', '–Ω–∞'])

#–ó–∞–º–µ–Ω—è—Ç—å —Ç–æ–ª—å–∫–æ —Å–ª–æ–≤–∞ –≤ —á–∏—Å—Ç–æ–º –≤–∏–¥–µ –Ω–µ —è–≤–ª—è—é—â–∏–µ—Å—è —á–∞—Å—Ç—å—é –¥—Ä—É–≥–æ–≥–æ —Å–ª–æ–≤–∞
bad_patterns = [f'(?<![a-z–∞-—è]){w}(?![a-z–∞-—è])'.format(w) for w in bad_words]
bad_patterns = '|'.join(bad_patterns)

#–û—Å—Ç–∞–ª—è—é —Ç–æ–ª—å–∫–æ —Å–ª–æ–≤–∞ –±–µ–∑ –≤—Å–µ—Ö —Å–∏–º–≤–æ–ª–æ–≤
p0 = '(?<=\W)\W|\W(?=\W)|(?<=^)\W|\W(?=$)'
p1 = '(?<=\w)-(?=\w)|\w'
p2 = '\W'

#–ó–∞–º–µ–Ω—ã, –±–∏–±–ª–∏–æ—Ç–µ–∫–∞ —Å —Ç—Ä–∞–Ω—Å–ª–∏—Ç–µ—Ä–∞—Ü–∏–µ–π –Ω–µ –ø–æ–¥–∫–ª—é—á–∞–µ—Ç—Å—è –ø–æ —ç—Ç–æ–º—É —Ä—É–∫–∞–º–∏.
replace_pairs = {'—Å–∞–±–≤–µ–π' : 'subway',
                 '—Å—Ç–∞—Ä–±–∞–∫—Å' : 'starbucks',
                 '–≤–æ–∫' : 'wok',
                 '–¥–∞–Ω–∫–∏–Ω –¥–æ–Ω–∞—Ç—Å': 'dunkin donuts',
                 '–∫–æ—Å—Ç–∞ –∫–æ—Ñ–µ' : 'costa coffee'}

#–§—É–Ω–∫—Ü–∏—é —Å–æ–≤—Å–µ–º –Ω–µ –æ–ø—Ç–∏–º–∞–ª—å–Ω–∞—è . . .
def clean_data(some_series):
    some_series = some_series.str.lower()
    some_series = some_series.apply(lambda x: re.sub(bad_patterns, '', x))
    some_series = some_series.str.replace(p2, ' ', regex=True).replace('  ', ' ', regex=True)
    
    for key, value in replace_pairs.items():  
        some_series = some_series.str.replace(f'(?<![a-z–∞-—è]){key}(?![a-z–∞-—è])'.format(key), value, regex=True)
    some_series = some_series.str.strip()
    return some_series
    
#–û–±—Ä–∞–±–∞—Ç—ã–≤–∞—é
all_chain_names = clean_data(all_chain_names)
tmp[1] = clean_data(tmp[1])

#–ß–∞—Å—Ç–æ—Ç–∞ –≤—Å—Ç—Ä–µ—á–∏ –Ω–∞–∑–≤–∞–Ω–∏–π
tmp = tmp.join(all_chain_names.value_counts(sort=False), on=tmp[1])
tmp.columns = [0, 1, 2]
```

#### –í—Ç–æ—Ä–æ–π —ç—Ç–∞–ø

–í –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è —á–∞—Å—Ç–æ—Ç—ã –≤—Å—Ç—Ä–µ—á–∏, –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–∞ –≤ 2 —Ä–∞–∑–∞ –±–æ–ª–µ–µ —á–∞—Å—Ç—ã–µ –µ—Å–ª–∏ –µ—Å—Ç—å –ø–æ–ª–Ω–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ –±–æ–ª–µ–µ —á–∞—Å—Ç–æ–≥–æ –∫–∞–∫ —Å–ª–æ–≤–æ.

```python
#tmp[3] = [list() for i in range(tmp.shape[0])]
#–°–ª–æ–≤–∞—Ä—å –∑–∞–º–µ–Ω, —á—Ç–æ –±—ã –ø–æ–ø—Ä–∞–≤–∏—Ç—å —Ñ—Ä–µ–π–º —Å–æ –≤—Å–µ–º–∏ –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ –∏ –ø–æ—Å—á–∏—Ç–∞—Ç—å —á–∞—Å—Ç–æ—Ç—É
replace_dict = dict()
#–ò—â—É –ø–æ–ª–Ω—ã–µ –≤—Ö–æ–∂–¥–µ–Ω–∏—è –Ω–∞–∑–≤–∞–Ω–∏–π –≤ –¥—Ä—É–≥–∏–µ –Ω–∞–∑–≤–∞–Ω–∏—è, –Ω–∞—á–∏–Ω–∞—è —Å —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö
i = 0
for index0, row0 in tmp.sort_values(by=2,ascending=False).iterrows():
    i+=1
    for index1, row1 in tmp.sort_values(by=2,ascending=False).iloc[i:].iterrows():
        #–ù–µ –æ–¥–Ω–æ –∏ —Ç–æ –∂–µ –Ω–∞–∑–≤–∞–Ω–∏–µ
        if row0[1] != row1[1]:
            if re.search(f'(?<![a-z–∞-—è]){row0[1]}(?![a-z–∞-—è])'.format(row0[1]), row1[1]) is not None:
                #–ï—Å–ª–∏ —á–∞—Å—Ç–æ—Ç–∞ –≤—Å—Ç—Ä–µ—á–∏ –Ω–∞–∑–≤–∞–Ω–∏—è –±–æ–ª—å—à–µ, —Ç–æ –ø–æ–≥–ª–∞—â–∞–µ–º
                if (row0[2] - row1[2]) > row1[2]:
                    replace_dict[row1[1]] = row0[1]
                    tmp.loc[index1, 1] = tmp.loc[index0, 1]
```

–¢–µ–ø–µ—Ä—å —Å–ª–µ–¥—É–µ—Ç –ø–µ—Ä–µ—Ä–∞—Å—á–∏—Ç–∞—Ç—å –∏–∑–º–µ–Ω–∏–≤—à—É—é—Å—è —á–∞—Å—Ç–æ—Ç—É –≤—Å—Ç—Ä–µ—á.

```python
#–ù–∞–¥–æ –ø–µ—Ä–µ—Å—á–∏—Ç–∞—Ç—å —Ä–µ–π—Ç–∏–Ω–≥–∏ –Ω–∞–∑–≤–∞–Ω–∏–π
all_chain_names = all_chain_names.replace(replace_dict)
del tmp[2]
tmp = tmp.join(all_chain_names.value_counts(sort=False), on=tmp[1])
tmp.columns = [0, 1, 2]
#telegram_raw = telegram_raw.join(tmp_df.loc[:,['id', 'desc_html']].set_index('id'), on='id')
```

–¢–∞–∫ –∂–µ –≤ –¥–∞–Ω–Ω—ã—Ö –µ—Å—Ç—å —Ä–µ–¥–∫–∏–µ —Å—Ç—Ä–æ–∫–∏ —Å –æ–¥–Ω–æ–π –∏ –±–æ–ª–µ–µ –æ–ø–µ—á–∞—Ç–∫–æ–π –≤ –Ω–∞–∑–≤–∞–Ω–∏–∏. –¢–µ —á—Ç–æ —Å –æ—Ç–ª–∏—á–∏–µ–º –≤ –æ–¥–∏–Ω —Å–∏–º–≤–æ–ª –º–æ–∂–Ω–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å.

```python
i=0
#–ò—â—É –æ—Ç–ª–∏—á–∞—é—â–∏–µ—Å—è –≤ –æ–¥–∏–Ω —Å–∏–º–≤–æ–ª –∏ –≤ –¥–≤–∞ —Ä–∞–∑–∞ –º–µ–Ω—å—à–∏–µ –ø–æ —Ä–µ–π—Ç–∏–≥—É –Ω–∞–∑–≤–∞–Ω–∏—è
#—Å–æ–º–Ω–∏—Ç–µ–ª—å–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ –Ω–æ –Ω–∞ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö —Ä–∞–±–æ—Ç–∞–µ—Ç
for index0, row0 in tmp.sort_values(by=2,ascending=False).iterrows():
    i+=1
    for index1, row1 in tmp.sort_values(by=2,ascending=False).iloc[i:].iterrows():
        #–û—Ç–ª–∏—á–∞—é—Ç—Å—è –≤ –æ–¥–∏–Ω —Å–∏–º–≤–æ–ª
        if nltk.edit_distance(row0[1], row1[1]) == 1:
            if (row0[2] - row1[2]) > row1[2]:
                tmp.loc[index1, 1] = tmp.loc[index0, 1]
```
